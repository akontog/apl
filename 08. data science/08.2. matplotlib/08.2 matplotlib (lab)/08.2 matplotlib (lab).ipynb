# Παράδοξο των γενεθλίων

Ένα κλασικό πρόβλημα πιθανοτήτων αποτελεί το λεγόμενο παράδοξο των γενεθλίων. Αφορά στο ερώτημα πόσοι άνθρωποι πρέπει να βρίσκονται σε ένα δωμάτιο ώστε η πιθανότητα να υπάρχουν τουλάχιστον δύο άτομα με τα ίδια γενέθλια να είναι μεγαλύτερη από 50%.
Για λόγους απλότητας δεν θα λάβουμε υπόψη δίσκετα έτη. 
Το πρώτο άτομο έχει γεννηθεί κάποια ημερομηνία από τις 365 πιθανές. Για να είναι διαφορετική η ημερομηνία του δεύτερου ατόμου θα πρέπει να έχει γεννηθεί μία ημερομηνία από τις 364 που απομένουν σε ένα έτος κ.ο.κ. Έτσι, η πιθανότητα να μην υπάρχουν κοινά γενέθλια n ατόμων είναι:
$$P(\text{όχι κοινά γενέθλια}) = \frac{{365}}{{365}} \times \frac{{364}}{{365}} \times \frac{{363}}{{365}} \times \ldots \times \frac{{365 - n + 1}}{{365}} = \frac{{365!}}{{365^n \times (365-n)!}}$$
Ουσιαστικά η πιθανότητα που ψάχνουμε είναι η συμπληρωματική της.
