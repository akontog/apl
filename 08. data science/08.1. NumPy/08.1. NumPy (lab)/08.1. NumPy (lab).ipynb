## Κινούμενος μέσος όρος

Να υλοποιηθεί συνάρτηση, έστω _movingAverage_, η οποία δέχεται διάνυσμα NumPy, _a_, και αριθμό, _w_, και επιστρέφει πίνακα με τον κινητό μέσο όρο του πίνακα _a_ με παράθυρο _w_.

Ο κινητός μέσος όρος για κάθε στοιχείο, εκτός των πρώτων και τελευταίων στοιχείων που δεν έχουν τα απαιτούμενα γειτονικά στοιχεία, θα υπολογίζεται ως ο μέσος όρος των w διαδοχικών αριθμών Ο w θα πρέπει να είναι περιττός και θετικός, όχι μεγαλύτερος από τον αριθμό στοιχείων του αρχικού πίνακα. Στο επόμενο παράδειγμα, ο τύπος δίνει το κάθε στοιχείο για w=3.
$$
MA(i) = \frac{a[i-1] + a[i] + a[i+1]}{3}
$$

Οι ακραίες τιμές του πίνακα, για τις οποίες δεν μπορεί να προκύψει μέσος όρος θα πρέπει να οριστούν ως `np.NaN`

Για τον πίνακα:
$$
a = [1,\ 5,\ 3,\ 6,\ 8,\ 2,\ 7]
$$
Ο πίνακας που θα προκύψει από την διαδικασία
$$
MA = [\text{NaN},\ 3.0,\ 4.67,\ 5.67,\ 5.33,\ 5.67,\ \text{NaN}]
$$
Παράδειγμα κλήσης:
```
>>>a = np.array([1, 5, 3, 6, 8, 2, 7])
>>>movingAverage(a, 3)

array([ nan, 3.  , 4.67, 5.67, 5.33, 5.67,  nan])
```

## Αποστάσεις

Να υλοποιηθούν συναρτήσεις python οι οποίες δέχονται σημεία ως tuples συντεταγμένων τα μετατρέπουν σε πίνακες NumPy και υπολογίζουν και επιστρέφουν τις επόμενες αποστάσεις.

1. Συνάρτηση _euclideanDistance_ για την ευκλείδεια απόσταση.
$$
d_{\text{Euclidean}}(A, B) = \sqrt{\sum_{i=1}^{n} (A_i - B_i)^2}
$$

2. Συνάρτηση _manhattanDistance_ για την απόσταση Μανχάταν.
$$
d_{\text{Manhattan}}(A, B) = \sum_{i=1}^{n} |A_i - B_i|
$$

3. Συνάρτηση _minkowskiDistance_ για την απόσταση Minkowski που αποτελεί και γενίκευση των παραπάνω αποστάσεων. Η συνάρτηση αυτή, πέρα από τα σημεία θα δέχεται και την παράμετρο p. Θα πρέπει για p=1 να επιστρέφει ίδια αποτελέσματα με την _manhattanDistance_ και για p=2 ίδια με την _euclideanDistance_. Η συνάρτηση θα πρέπει να έχει εξορισμού τιμή p=3.
$$
d_{\text{Minkowski}, p}(A, B) = \left( \sum_{i=1}^{n} |A_i - B_i|^p \right)^{\frac{1}{p}}
$$

Στην συνέχεια απεικονίζονται οι αποστάσεις για τις παραπάνω περιπτώσεις
$$
\begin{array}{c c c}
\textbf{Ευκλείδεια} & \textbf{Manhattan} & \textbf{Minkowski (p=3)} \\ 
\hline
\begin{bmatrix} 
\sqrt{2} & 1 & \sqrt{2} \\ 
1 & 0 & 1 \\ 
\sqrt{2} & 1 & \sqrt{2} 
\end{bmatrix} 
&
\begin{bmatrix} 
2 & 1 & 2 \\ 
1 & 0 & 1 \\ 
2 & 1 & 2 
\end{bmatrix} 
&
\begin{bmatrix} 
1.26 & 1 & 1.26 \\ 
1 & 0 & 1 \\ 
1.26 & 1 & 1.26 
\end{bmatrix} 
\end{array}
$$

Παράδειγμα κλήσης:
```
>>>A = (1, 2)
>>>B = (4, 6)
>>>euclideanDistance(A,B)

5.0
>>>manhattanDistance(A,B)

7
>>>minkowskiDistance(A,B)

4.497941445275415
```

## Μαγικό τετράγωνο

Το μαγικό τετράγωνο αποτελεί μια διάταξη μοναδικών αριθμών σε τετραγωνικό πίνακα, όπου η αριθμητική πράξη μεταξύ των αριθμών στην ίδια σειρά, στήλη και διαγώνιο του τετραγώνου επιστρέφει πάντα το ίδιο αποτέλεσμα. Αν η πράξη είναι η πρόσθεση, το αποτέλεσμα για ένα μαγικό τετράγωνο $n \times n$ δίνεται από τον τύπο $M = \frac{n(n^2 + 1)}{2}$. Επιπλέον οι αριθμοί που θα περιέχει το τετράγωνο θα είναι από 1 έως $n^2$.

Ένα μαγικό τετράγωνο 3×3 με M = 15:

$$
\begin{array}{|c|c|c|}
\hline
8 & 1 & 6 \\
\hline
3 & 5 & 7 \\
\hline
4 & 9 & 2 \\
\hline
\end{array}
$$

1. Υλοποιήστε συνάρτηση, έστω readSquare(filename), η οποία δέχεται αρχείο με αριθμούς χωρισμένους με ',' όπως στην συνέχεια και επιστρέφει πίνακα NumPy.
```
8,1,6  
3,5,7  
4,9,2
```

Παράδειγμα κλήσης:
```
>>>readSquare("magic_true.txt")

array([[8, 1, 6],
       [3, 5, 7],
       [4, 9, 2]])
```

2. Υλοποιήστε συνάρτηση, έστω isMagic(square), που δέχεται έναν πίνακα NumPy και επιστρέφει True αν αποτελεί μαγικό τετράγωνο. Διαφορετικά επιστρέφει False.

Παράδειγμα κλήσης:
```
>>>a = readSquare("magic_true.txt")
>>>isMagic(a)

True

>>>b = readSquare("magic_false.txt")
>>>isMagic(b)

False
```
