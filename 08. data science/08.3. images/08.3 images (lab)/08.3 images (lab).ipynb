## Στεγανογραφία

Η στεγανογραφία, steganography, προέρχεται από τις λέξεις στεγανός και γραφή και αποτελεί τρόπο απόκρυψης της ύπαρξης ενός μηνύματος. Σε αντίθεση με την κρυπτογραφία, που προστατεύει το περιεχόμενο ενός μηνύματος, η στεγανογραφία προστατεύει το γεγονός ότι υπάρχει μήνυμα. Ενσωματώνει πληροφορία μέσα σε φαινομενικά αθώα δεδομένα – όπως εικόνες, ήχους ή βίντεο – χωρίς να είναι ορατή η αλλοίωση στον παρατηρητή.

1. Υλοποιήστε συνάρτηση, έστω encodeStr η οποία δέχεται ένα μήνυμα, έστω message, σε μορφή `str` και επιστρέφει `str` με 8bit για κάθε χαρακτήρα του μηνύματος. Υπενθυμίζεται ότι η `ord` επιστρέφει τον κωδικό ASCII του χαρακτήρα και η `bin` μετατρέπει δεκαδικό σε δυαδικό. 

Προσοχή: 

    - η `bin` επιστρέφει και την ένδειξη '0b' μπροστά που θα πρέπει να αφαιρέσουμε.
    
    - επιπλέον, τα bit δεν είναι πάντα σταθερός αριθμός. Θα πρέπει να προσθέσουμε μηδενικά αριστερά ώστε κάθε χαρακτήρας να καταλαμβάνει 8 bit.
    
Παράδειγμα κλήσης:

```
>>>encodeStr("Hello")

'0100100001100101011011000110110001101111'
```

Παράδειγμα κλήσης:

```
>>>decodeStr('0100100001100101011011000110110001101111')
```
