# Κλάσεις και Αντικείμενα, Classes and Objects

## Person

Δημιουργήσετε κλάση **Person** με τις εξής ιδιότητες/μεθόδους:

- Ιδιότητες:

    - _id_

    - _lastname_

    - _firstname_

    - _yearofbirth_

- Μέθοδοι:

    - κατασκευαστής της κλάσης ο οποίος δέχεται και αρχικοποιεί τις παραπάνω ιδιότητες.

    - _getAge_ η οποία δέχεται ως παράμετρο το τρέχον έτος και επιστρέφει την ηλικία του συγκεκριμένου προσώπου.

    - _display_, η οποία θα εκτυπώνει τα στοιχεία του προσώπου (id, Ονοματεπώνυμο, έτος γέννησης).

Να γραφεί πρόγραμμα σε Python, το οποίο θα δημιουργεί δύο αντικείμενα της κλάσης _Person_ και θα καλεί τις μεθόδους _display_, _getAge_ για να εκτυπώνει τα στοιχεία και την ηλικία του προσώπου.

Ένα παράδειγμα χρήσης της παραπάνω κλάσης για ένα πρόσωπο είναι το εξής:

```
>>> student1 = Person('id-001', 'John', 'Doe', 1995)

>>> student1.display()

ID: id-001
Ονοματεπώνυμο: John Doe
Έτος γέννησης: 1995
Ηλικία: 30

>>> print("Ηλικία:", student1.getAge(2025))

Ηλικία: 30
>>>
```

## *Εφαρμογή, Κρυπτογράφημα του Καίσαρα

Υλοποιήσετε κλάση _Cipher_, η οποία θα χρησιμοποιείται για την κρυπτογράφηση και αποκρυπτογράφηση κεφαλαίων χαρακτήρων του Λατινικού αλφαβήτου με βάση την τεχνική της μετατόπισης χαρακτήρων, Caesar Cipher.
Η κλάση θα έχει τις εξής ιδιότητες και μεθόδους:

- Ιδιότητες:

    - _shift_: Η τιμή της μετατόπισης που θα χρησιμοποιείται για την κρυπτογράφηση και αποκρυπτογράφηση. Θα ορίζεται κατά τη δημιουργία του αντικειμένου.

    - _res_: Λίστα που θα αποθηκεύει τα κρυπτογραφημένα δεδομένα. Αρχικά θα είναι άδεια.

- Μέθοδοι:

    - _encrypt_: Δέχεται τα προς κρυπτογράφηση δεδομένα σε μορφή `str` και τα αποθηκεύει στην λίστα _res_ ως λίστα κρυπτογραφημένων χαρακτήρων. Προσοχή, αν για παράδειγμα ο χαρακτήρας είναι ο 'Y' και η μετατόπιση _shift_ έχει τεθεί 2, ο χαρακτήρας πρέπει να κρυπτογραφηθεί σε 'A'.

    - _decrypt_: Επιστρέφει τα περιεχόμενα της _res_ σε μορφή `str` με τα δεδομένα αποκρυπτογραφημένα

    - _display_: Εμφανίζει τα περιεχόμενα της λίστας _res_.

Υλοποιήστε πρόγραμμα το οποίο θα δημιουργεί ένα αντικείμενο της _Cipher_ με μετατόπιση της επιλογής σας. Στην συνέχεια θα κρυπτογραφεί μία φράση, έστω την "THE QUICK BROWN FOX JUMPS OVER THE LAZY DOG".

Παράδειγμα:
```
>>>cipher = Cipher(shift=2)
>>>phrase = "THE QUICK BROWN FOX JUMPS OVER THE LAZY DOG"
>>>cipher.encrypt(phrase)
>>>cipher.display()

VJG SWKEM DTQYP HQZ LWORU QXGT VJG NCBA FQI
>>>cipher.decrypt()

THE QUICK BROWN FOX JUMPS OVER THE LAZY DOG
```

## Τράπεζα

Υλοποιήστε ένα σύστημα διαχείρισης τραπεζικών λογαριασμών το οποίο περιλαμβάνει τις κλάσεις με τις ιδιότητες και μεθόδους που ακολουθούν:

- _BankAccount_:

    - Ιδιότητες:

        - _id_ μοναδικό αναγνωριστικό που θα προκύπτει αυτοματοποιημένα από μία static μεταβλητή μετρητή λογαριασμών

        - _holders_: λίστα με id κατόχων η οποία αρχικά θα είναι άδεια.

        - _balance_: υπόλοιπο λογαριασμού.

    - Μέθοδοι:
    
        - `__init__`: Μέθοδος κατασκευαστή. Δέχεται ένα αρχικό ποσό, εξορισμού 0, και θέτει κενή λίστα κατόχων του λογαριασμού. 

        - _deposit_: κατάθεση ενός ποσού. Η κατάθεση επιτρέπεται μόνο αν το ποσό είναι θετικό. Επιστρέφει True ή False, ανάλογα.

        - _withdraw_: ανάληψη ποσού. Η ανάληψη επιτρέπεται μόνο όταν υπάρχει αρκετό υπόλοιπο. Επιστρέφει True ή False, ανάλογα.

        - _getBalance_: επιτρέφει το τρέχον υπόλοιπο.

        - _display_: εμφάνιση στοιχείων λογαριασμού. Θα εμφανίζει το μοναδικό id, την λίστα με τα id των κατόχων και το τρέχων υπόλοιπο.

- _SavingsAccount_: Λογαριασμός ταμιευτηρίου. Κληρονομεί από την _BankAccount_
    - Ιδιότητες:
       
        - _interest_: το επιτόκιο του λογαριασμού ταμιευτηρίου, εξορισμού 0.02.

    - Μέθοδοι:
    
        - `__init__`: Δέχεται αρχικό ποσό, εξορισμού 0, και το επιτόκιο του λογαριασμού ταμιευτηρίου. Καλεί κατάλληλα τον κατασκευαστή της υπερκλάσης.
    
        - _applyInterest_: Εφαρμόζει το επιτόκιο στο υπόλοιπο του λογαριασμού.
        
        - _display_: Εμφανίζει το είδος του λογαριασμού, δηλαδή "Savings Account" και καλεί την _display_ της υπερκλάσης για τα υπόλοιπα στοιχεία. Τέλος εμφανίζει το επιτόκιο.
 
 - _CheckingAccount_: Τρεχούμενος λογαριασμός, κληρονομεί από την _BankAccount_.

    - Ιδιότητες:

        - _creditlimit_: Πιστωτικό όριο του λογαριασμού, δηλαδή το ποσό υπερανάληψης του.

    - Μέθοδοι:
    
        - `__init__`: Δέχεται αρχικό ποσό, εξορισμού 0, και το πιστωτικό όριο. Καλεί κατάλληλα τον κατασκευαστή της υπερκλάσης.
    
         - _withdraw_: ανάληψη ποσού. Η ανάληψη επιτρέπεται μόνο όταν υπάρχει αρκετό υπόλοιπο. Το υπόλοιπο μπορεί να γίνει και αρνητικό μέχρι και το πιστωτικό όριο του λογαριασμού. Επιστρέφει True ή False, ανάλογα.

        - _display_: Εμφανίζει το είδος του λογαριασμού, δηλαδή "Checking Account" και καλεί την _display_ της υπερκλάσης για τα υπόλοιπα στοιχεία. Τέλος εμφανίζει το πιστωτικό όριο.
        
- _Customer_:

    - Ιδιότητες:

        - _id_: μοναδικό αναγνωριστικό που προκύπτει αυτοματοποιημένα με τη βοήθεια static μεταβλητής.

        - _accounts_: λίστα λογαριασμών του πελάτη, η οποία αρχικά θα είναι άδεια.

        - _fullname_: όνομα του πελάτη.

    - Μέθοδοι:

        - _addAccount_ για προσθήκη λογαριασμού στη λίστα λογαριασμών.

        - _display_ για εμφάνιση στοιχείων πελάτη.

- _Bank_:

    - Ιδιότητες:
        - _customers_: λεξικό πελατών με κλειδί το id του πελάτη και τιμή αντικείμενο πελάτη.

        - _accounts_: λεξικό λογαριασμών με κλειδί το id του λογαριασμού και τιμή αντικείμενο λογαριασμού.

    - Μέθοδοι:

        - _addCustomer_: Θα δέχεται παράμετρο το όνομα του πελάτη και θα δημιουργεί αντικείμενο πελάτη που θα αποθηκεύει κατάλληλα στο λεξικό πελατών. Τέλος, θα επιστρέφει το αντικείμενο του πελάτη που δημιούργησε.

            - _addAccount_: Θα δημιουργεί αντικείμενο λογαριασμού που θα αποθηκεύει κατάλληλα στο λεξικό λογαριασμών. Θα επιστρέφει το αντικείμενο του λογαριασμού που δημιούργησε. Θα δέχεται τις παραμέτρους που ακολουθούν:
                - το αρχικό υπόλοιπο του λογαριασμού που εξορισμού θα είναι μηδενικό, 
                - το είδος του λογαριασμού που μπορεί να είναι εξορισμού 'BankAccount' ή διαφορετικά "SavingsAccount", "CheckingAccount".
                - το creditlimit, εξορισμού None, σε περίπτωση τρεχούμενου λογαριασμού.
                - το interest, εξορισμού None, σε περίπτωση λογαριασμού ταμιευτηρίου.
                
        - _addAccountToCustomer_: για προσθήκη λογαριασμού σε πελάτη. Θα δέχεται id λογαριασμού και πελάτη θα βρίσκει στα αντίστοιχα λεξικά τα αντικείμενα. Θα προσθέτει κατάλληλα στη λίστα κατόχων του λογαριασμού τον πελάτη και στη λίστα λογαριασμών του πελάτη τον λογαριασμό.

        - _display_: εμφάνιση όλων των πελατών και λογαριασμών.

# Αφηρημένες, abstract, κλάσεις και μέθοδοι

## Police and Thief

Το παιχνίδι Police and Thief περιλαμβάνει δύο βασικούς ήρωες, τον «Αστυνόμο», P, και τον «Κλέφτη», T, που κινούνται σε ένα ταμπλό διάστασης NxN. Ο αστυνόμος ελέγχεται από τον χρήστη και κινείται κατά ένα βήμα σε κάθε γύρο, ενώ ο κλέφτης ελέγχεται από τον υπολογιστή κινούμενος τυχαία κατά δύο βήματα. Το παιχνίδι τελειώνει όταν ο Αστυνόμος πιάσει τον κλέφτη ή όταν περάσει κάποιος αριθμός γύρων.
Ακολουθεί παράδειγμα ταμπλό του παιχνιδιού με την ένδειξη P να δείχνει την θέση του Αστυνόμου και την T τον κλέφτη.
```
- - - - - - - - - -
- - - - - - - - - -
- - - - - - - - - -
- - P - - - - - - -
- - - - - - - - - -
- - - - - - - - - -
- - - - - - - - - -
- - - - - T - - - -
- - - - - - - - - -
- - - - - - - - - -
```
Θα χρειαστούν οι επόμενες κλάσεις:

1. **Character**: Η κλάση θα είναι αφηρημένη, δεν θα μπορούμε να δημιουργήσουμε αντικείμενά της.

    - Ιδιότητες του τις συντεταγμένες (_x_, _y_) που βρίσκεται ο χαρακτήρας καθώς και το σύμβολό του, _symbol_. Οι παραπάνω ιδιότητες τίθενται με την αρχικοποίηση στον κατασκευαστή.

    - Η μέθοδος _move_ δέχεται την κατεύθυνση που θα κινηθεί ο χαρακτήρας, μία τιμή, δηλαδή, μεταξύ των 'up', 'down', 'left', 'right', και ένα `tuple` με το μέγεθος του ταμπλό. Επιτρέπει την κίνηση του ήρωα σε μία από τις τέσσερις κατευθύνσεις, ελέγχοντας τα όρια. Αν για παράδειγμα βρίσκεται στην τελευταία γραμμή του ταμπλό και η εντολή κατεύθυνσης είναι  'down' ο χαρακτήρας θα παραμείνει ακίνητος.

2. **Police**:

    - Υποκλάση της _Character_ που αναπαριστά τον Αστυνόμο με το σύμβολο 'P'.

    - Περιέχει τη μέθοδο _move_ που επιτρέπει την κίνηση του αστυνόμου ζητώντας την είσοδο του χρήστη και καλώντας κατάλληλα την _move_ της _Character_.

3. **Thief**:

    - Υποκλάση της _Character_ που αναπαριστά τον κλέφτη με το σύμβολο 'T'.

    - Περιέχει τη μέθοδο _move_ που κινεί τον κλέφτη τυχαία κατά δύο βήματα καλώντας κατάλληλα δύο φορές την _move_ της _Character_.

4. **Game**:

    - Μέθοδος κατασκευαστή που δέχεται ακέραιο _n_ με το μέγεθος του ταμπλό και το αρχικοποιεί με τις θέσεις των ηρώων, αστυνόμου και κλέφτη. Οι ήρωες θα πρέπει να τοποθετούνται σε τυχαία θέση που όμως δεν θα μπορεί να είναι η ίδια.

    - Μέθοδος _update_board_ η οποία καταχωρεί τον χαρακτήρα '-' σε όλα τα στοιχεία του ταμπλό εκτός από τα σημεία που βρίσκονται κλέφτης και αστυνόμος όπου καταχωρεί 'T' και 'P' αντίστοιχα.

    - Μέθοδος _display_ εμφανίζει την πίστα όπως φαίνεται και στο παράδειγμα παραπάνω.

    - Η μέθοδος _play_ διαχειρίζεται την κύρια επανάληψη του παιχνιδιού,  καλώντας κατάλληλα την _move_ του κάθε παίκτη και ελέγχοντας σε κάθε γύρο αν ο αστυνόμος έχει "πιάσει" τον κλέφτη, αν δηλαδή βρίσκονται στις ίδιες συντεταγμένες του ταμπλό. Μπορείτε να αλλάξετε και την ένδειξη της κοινής συντεταγμένης για παράδειγμα σε 'X'. Η επανάληψη σταματά μετά από προκαθορισμένο αριθμό γύρων, έστω rounds = 10,  ή όταν πιαστεί ο κλέφτης.

## Βιβλιοπωλείο

Θα πρέπει να υλοποιηθεί πρόγραμμα για διαχείριση μίας αποθήκης για χρήση από βιβλιοπωλεία. Θα απαιτηθούν οι κλάσεις που ακολουθούν:

1.	Κλάση **Product** που θα αναπαριστά ένα προϊόν. Σημειώνεται ότι το βιβλιοπωλείο εκτός από βιβλία μπορεί να έχει γραφική ύλη παιχνίδια κ.λπ.. Ένα προϊόν, _Product_ δεν θα μπορεί να δημιουργηθεί ως αντικείμενο, θα πρέπει να είναι αφηρημένη κλάση.

    - Ιδιότητες:

        - _id_: πρέπει να είναι αύξον αριθμός που δίνεται μοναδικά σε κάθε προϊόν αυτοματοποιημένα με χρήση βοηθητικής static μεταβλητής.

        - _title_: Τίτλος του προϊόντος.

        - _price_: Τιμή του.

        - _stock_: Απόθεμά του στο βιβλιοπωλείο.

    - Μέθοδοι:

        - `__init__`: Κατασκευαστής, για δημιουργία ενός προϊόντος με αρχικοποίηση των παραπάνω στοιχείων.

        - _display_: Αφηρημένη μέθοδος εμφάνισης πληροφοριών.

2. Κλάση **Book** που θα αναπαριστά ένα βιβλίο το οποίο **είναι και προϊόν** με:

    - Ιδιότητες:

        - _author_: συγγραφέας, μία συμβολοσειρά με το ονοματεπώνυμο του συγγραφέα.

        - _year_: έτος έκδοσης του βιβλίου.

        - _genre_: η κατηγορία του βιβλίου, συμβολοσειρά.

        - _pages_: ο αριθμός σελίδων του.

    - Μέθοδοι:

        - `__init__`: Κατασκευαστής,  για δημιουργία ενός βιβλίου με αρχικοποίηση των παραπάνω στοιχείων του βιβλίου που όμως είναι και προϊόν. Έτσι θα καλεί κατάλληλα και τον κατασκευαστή της _Product_.

        - _display_: Κατάλληλη εμφάνιση πληροφοριών βιβλίου.

3.	Κλάση **Toy** που θα αναπαριστά ένα παιχνίδι το οποίο **είναι προϊόν** με:

    - Ιδιότητες:

        - _kind_: Είδος παιχνιδιού, συμβολοσειρά.

     - Μέθοδοι:

        - `__init__`: Κατασκευαστής,  για δημιουργία ενός παιχνιδιού. Θα αρχικοποιείται η ιδιότητα του παιχνιδιού και θα καλείται κατάλληλα και ο κατασκευαστής της _Product_.

        - _display_: Κατάλληλη εμφάνιση πληροφοριών παιχνιδιού.

4.	Κλάση **Bookstore** που θα αναπαριστά ένα βιβλιοπωλείο και θα διαχειρίζεται βιβλία και άλλα παιχνίδια και παραγγελίες:

    - Ιδιότητες:

        - _products_: Λεξικό

    - Μέθοδοι:

## *MyComplex

Δημιουργήστε κλάση, **MyComplex**, για μιγαδικούς αριθμούς χωρίς να χρησιμοποιήσετε τον αντίστοιχο τύπο της γλώσσας Python. Η κλάση θα περιλαμβάνει:

    - ιδιότητες:

        - _r_: πραγματικό μέρος

        - _im_: φανταστικό μέρος

    - μεθόδους:

        - `__init__`: κατασκευαστής, ο οποίος θα δημιουργεί μιγαδικούς με πραγματικό και φανταστικό μέρος. Όταν το φανταστικό μέρος δεν θα δίνεται, θα είναι εξορισμού το 0.

        - `__str__`: επιστρέφει μια φιλική αναπαράσταση του μιγαδικού αριθμού, π.χ., '3 + 4i'.

        - `__add__` και `__sub__`: θα δέχονται παράμετρο _z_ και θα τον προσθέτουν ή αφαιρούν αντίστοιχα στο αντικείμενο στο οποίο εφαρμόζονται επιστρέφοντας νέο μιγαδικό αριθμό.

        - `__mul__`: θα δέχεται παράμετρο _z_ και θα υπολογίζει και θα επιστρέφει το γινόμενό του με το αντικείμενο εφαρμογής.

        - `__truediv__`: θα δέχεται παράμετρο _z_ και θα υπολογίζει και θα επιστρέφει το πηλίκο του με το αντικείμενο εφαρμογής.

        - `__eq__`: Ελέγχει την ισότητα μεταξύ δύο μιγαδικών αριθμών. Επιστρέφει `True` ή `False` ανάλογα.

        - `__neg__`: Επιστρέφει τον αρνητικό του μιγαδικού αριθμού, π.χ., αν z = 3+4i, τότε -z = -3-4i.

        - `__abs__`: Υπολογίζει το μέτρο του μιγαδικού αριθμού.

Έστω μιγαδικοί αριθμοί $z_1$=a+bi, $z_2$=c+di

Γινόμενο: $z_1 \cdot z_2 = (a + bi)(c + di) = (ac - bd) + (ad + bc)i$

Πηλίκο: $z_1 / z_2 = \frac{(ac + bd) + (bc - ad)i}{c^2 + d^2}$

Μέτρο μιγαδικού αριθμού z=a+bi: $|z| = \sqrt{a^2 + b^2}$

Παράδειγμα εκτέλεσης:
```
>>>z1 = MyComplex(3, 4)
>>>z2 = MyComplex(1, -2)

>>>z1 + z2

4 + 2i

>>>z1 - z2

2 + 6i

>>>z1 * z2

11 - 2i

>>>z1 / z2

-1.0 + 2.0i

>>>z1 == z2

False

>>>-z1

-3 - 4i

>>>abs(z1)

5.0
```
