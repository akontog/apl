Σε άλλες περιπτώσεις η σύνταξη είναι σωστή αλλά καταλήγουμε σε μη αναμενόμενο αποτέλεσμα. Σε αυτές τις περιπτώσεις έχουμε κάνει λογικό λάθος. Άλλες φορές σφάλματα προκύπτουν μη αναμενόμενα κατά την ώρα της εκτέλεσης με αποτέλεσμα το πρόγραμμα να τερματίσει. Για παράδειγμα η μή εύρεση ενός αρχείου ή δικτύου, ή ακόμη και η προσπάθεια διαίρεσης με το 0. Πρόκειται για λογικά λάθη.

 Η Python χρησιμοποιεί εντολές try για τον χειρισμό λαθών (εξαιρέσεις) στον κώδικα. Με αυτόν τον τρόπο, το πρόγραμμα δεν τερματίζει. Η ροή του προγράμματος από το σημείο που συνέβη η εξαίρεση πηγαίνει στην except που ταιριάζει σε αυτήν.

```python
try:
    number1 = 23
    number2 = 0
    result = number1/number2
    print('Επιτυχής πράξη') # Εάν παραπάνω προκύψει εξαίρεση, η ροή δεν θα φτάσει ποτέ εδώ
except ZeroDivisionError:
    print('Ο παρονομαστής πρέπει να είναι διάφορος από το 0')

    
```

```python
try:
    with open('accounts0.txt', 'r') as accounts:
        for item in accounts:
            print(item)
except FileNotFoundError:
    print("Το αρχείο δεν υπάρχει!")
```

Πολύ συχνά σε ένα τμήμα κώδικα, ειδικά αν ζητείται είσοδος από τον χρήστη, μπορεί να είναι πιθανό να προκύψουν παραπάνω από ένας τύπος εξαίρεσης. Οι τύποι αυτοί μπορούν να ελεγχθούν διαδοχικά. Επίσης, με την <b>else</b> μετά τις except μπορούμε να θέσουμε κώδικα που θέλουμε να εκτελεστεί αν δεν προκύψει καμία εξαίρεση στο block Κώδικα που βρίσκεται στην try.

```python
try:
    number1 = int(input("Δώσε 1ο ακέραιο:"))
    number2 = int(input("Δώσε 1ο ακέραιο:"))
    result = number1/number2
except ValueError:
    print('Πρέπει να δώσετε ΜΟΝΟ ακέραιες τιμές:\n')
except ZeroDivisionError:
    print('Ο παρονομαστής πρέπει να είναι διάφορος από το 0')
else:
    print(f'Αποτέλεσμα: {result:3f}')
```

Είναι πιθανό σε κώδικα σε ένα block try να προκύψει κάποια εξαίρεση αλλά να μην γίνεται χειρισμός της εξαίρεσης αυτής. Τότε η εκτέλεση σταματά. Στο επόμενο παράδειγμα ο τύπος της εξαίρεσης είναι NameError αφού δεν έχει γίνει δηλωθεί η μεταβλητή y αλλά ο χειρισμός αφορά μόνο την διαίρεση με το 0.

```python
try:
    print(2 / y)
except ZeroDivisionError:
    print('Ο παρονομαστής πρέπει να είναι διάφορος από το 0')
```

Παρατηρήσαμε σε προηγούμενο παράδειγμα ότι σε περίπτωση που δεν συμβεί κάποια εξαίρεση μπορούμε να εκτελέσουμε ένα block κώδικα με χρήση της else. Υπάρχουν περίπτώσεις που κάποιο block κώδικα θέλουμε να εκτελεστεί σε οποιαδήποτε περίπτωση. Τότε γίνεται χρήση της <b>final</b>. Στο επόμενο παράδειγμα δοκιμάστε να δώσετε μή ακέραια τιμή, μηδενική τιμή και ακέραια τιμή διάφορη του 0 και παρατηρήστε. 
<ul>
    <li><b>Είσοδος 0</b>:Ο χειρισμός της μηδενικής τιμής δεν έχει γίνει με αποτέλεσμα το πρόγραμμα να τερματίζει, το final όμως εκτελείται.</li>
    <li><b>Είσοδος μή ακέραιος</b>:Υπάρχει χειρισμός για την ValueError με αποτέλεσμα η ροή του προγράμματος να συνεχίζει κανονικά. Εμφανίζεται πέρα από το μήνυμα για το ValueError και το μήνυμα στο final αλλά και το μήνυμα μετά την try.</li>
    <li><b>Είσοδος ακέραιος, μη μηδενικός</b>: Δεν προκύπτει κάποια εξαίρεση, εμφανίζεται το αποτέλεσμα, τα μηνύματα σε else, finally αλλά και στο τέλος μετά την try</li>
   </ul>

```python
try:
    d = int(input('Δώσε ακέραια τιμή:'))
    print(10/d)
except ValueError:
    print('Πρέπει να δώσετε ΜΟΝΟ ακέραια τιμή.')
else:
    print('Όλα πήγαν καλά')
finally:
    print('Αυτο το μήνυμα θα τυπωθεί όπως και να έχει.')
print('Όλα τελείωσαν ομαλά')
```

Στα παραπάνω παραδείγματα έγινε χρήση κάποιων ειδών εξαιρέσεων (FileNotFoundError, ValueError, ZeroDivisionError, NameError). Στην συνέχεια παρουσιάζονται διάφορες χρήσιμες εξαιρέσεις που συναντάμε συχνά:

    <li><b>IndexError</b>:Ο δείκτης εκτός ορίων σε μία δομή όπως η λίστα.</li>
    <li><b>TypeError</b>:Μία συνάρτηση ή ενσωματωμένη λειτουργία χρησιμοποιήθηκε σε αντικείμενο που δεν την υποστηρίζει.</li>
    <li><b>AttributeError</b>:Προσπάθεια αναφοράς σε χαρακτηριστικό που δεν υπάρχει.</li>
    <li><b>FileNotFoundError</b>:Το αρχείο δεν υπάρχει.</li>

Σημειώνεται ότι η <b>Exception</b> αποτελεί την υπερκλάση από την οποία κληρονομούν όλες οι εξαιρέσεις. Σε περιπτώσεις που υπάρχουν πολλές πιθανές εξαιρέσεις ή δεν γνωρίζουμε ποια εξαίρεση μπορεί να προκύψει μπορούμε να θέσουμε except: και να να πιαστεί οποιαδήποτε εξαίρεση, κάτι που δεν θεωρείται σωστή πρακτική. Επιπλέον, σε μία μόνο except μπορούμε να θέσουμε ένα tuple εξαιρέσεων σε μία μόνο except.

```python
try:
    L = [1,2,'3']
    print(L[1]+L[2]) # TypeError unsupported operand type(s) for +: 'int' and 'str'
    print(L[2]+L[3]) # IndexError: list index out of range
except:
    print('Προέκυψε κάποιο σφάλμα')
try:
    L = [1,2,'3']
    print(L[1]+L[2]) # TypeError unsupported operand type(s) for +: 'int' and 'str'
    print(L[2]+L[3]) # IndexError: list index out of range
except (TypeError,IndexError):
    print('Προέκυψε είτε TypeError είτε IndexError')
```

Το εξορισμού except πρέπει να είναι τελευταίο μετά από τα εξεδικευμένα.

```python
try:
    L = [1,2,'3']
    print(L[1]+L[2]) # TypeError unsupported operand type(s) for +: 'int' and 'str'
    print(L[2]+L[3]) # IndexError: list index out of range
except:
    print('Προέκυψε κάποιο error')
except TypeError:
    print('Προέκυψε TypeError')
except IndexError: 
    print('Προέκυψε IndexError')
```

Ο προγραμματιστής μπορεί να δημιουργήσει κώδικα που προκαλεί εξαιρέσεις ή και να δημιουργήσει δικά του είδη εξαιρέσεων.

```python
def square(n):
    if not isinstance(n, int |float):
        raise TypeError('Η παράμετρος θα πρέπει να είναι αριθμητικό δεδομένο')
    return n**2
square(42)
square('42')
```

```python
class GradeError(Exception):
    pass
# Υπολογίζει τον μέσο όρο των βαθμών
def getAvg(grades):
    total = 0
    for g in grades:
        if g < 0 or g > 10:
            raise GradeError('Οι βαθμοί πρέπει να είναι στο διάστημα [0,10]')
        total +=g
    return total/len(grades)
print(getAvg([5,6,5,5.5,7]))
print(getAvg([5,6,15,5.5,7]))
```

## Είσοδος και έξοδος με δυαδικά αρχεία

Η εργασία με αρχεία κειμένου στο πλαίσιο της numpy είναι σκόπιμη όταν ο όγκος των δεδομένων δεν είναι ιδιαίτερα μεγάλος ή όταν θέλουμε να έχουμε μια εναλλακτική εικόνα των δεδομένων, π.χ. με τη χρήση προγραμμάτων λογιστικών φύλλων (Excel, LibreOffice, κλπ). 

Σε περιπτώσεις μεγάλου όγκου δεδομένων η εργασία γίνεται με *δυαδικά* αρχεία με τη χρήση των συναρτήσεων`load()` και `save()` της numpy.


Η συνάρτηση `save()` επιτρέπει την αποθήκευση ενός πίνακα numpy σε δυαδική μορφή. Το αρχείο αποθηκεύεται σε μορφή `.npy`, όπως φαίνεται στο παράδειγμα:

```python

a = np.arange(100)

np.save("afile.npy",a)
```

```python
b = np.load("afile.npy")
print(b)
```

Εναλλακτικά, είναι δυνατόν να διαχειριστούμε αρχεία της numpy τα οποία ανοίγουν με κατάλληλη χρήση της συνάρτησης `open()`. Στην περίπτωση αυτή, οι `load()` και `save()` δεν δέχονται ως παράμετρο το όνομα του αρχείου, αλλά ένα αντικείμενο αρχείου, όπως στο παράδειγμα:

```python
with open('test.npy', 'wb') as f:
    np.save(f, np.array([1, 2]))

    np.save(f, np.array([1, 3]))

with open('test.npy', 'rb') as f:

    a = np.load(f)

    b = np.load(f)

print(a, b)
```

## Ανίχνευση περιγράμματος με εφαρμογή του λαπλασιανού τελεστή

Ο τελεστής Laplace ορίζεται ως:
$$\nabla ^2 f(x,y) = \frac{\partial ^2  f(x,y)}{\partial x^2} +\frac{\partial ^2 f(x,y)}{\partial y^2}$$

Ο τελεστής εφαρμόζεται πάνω σε μια παραγωγίσιμη συνάρτηση, $f(x,y)$ δύο μεταβλητών.

Σε διακριτή μορφή ο τελεστής είναι:

$$\Delta f(i,j) = \frac{f(i+1,j) + f(i-1,j) + f(i,j+1) + f(i,j-1) -4 f(i,j)}{h}$$

όπου $f(i,j)$ η τιμή της συνάρτησης στα σημεία $(i,j)$ ενός ορθογώνιου πλέγματος οριζόντιας και κατακόρυφης απόστασης $h$.

Ο διακριτός τελεστής μπορεί να εφαρμοστεί για την ανίχνευση περιγραμμάτων σε μια ψηφιακή εικόνα. Η συνάρτηση $f$ περιγράφει τη φωτεινότητα του εικονοστοιχείου (pixel) $i,j$.

Στην εφαρμογή που ακολουθεί εφαρμόζουμε τον τελεστή πάνω στα στοιχεία ενός πίνακα numpy ο οποίος περιέχει τις φωτεινότητες των εικονοστοιχείων μιας εικόνας. 

Η υλοποίηση του τελεστή στη numpy δίνεται ως εξής: 

```python
def laplacian(array):
    n = array.shape[0] # Πρώτη διάσταση (γραμμές)
    m = array.shape[1] # Δεύτερη διάσταση (στήλες)

    d = np.zeros((n,m))
    for i in range(1,n-1):
        for j in range(1,m-1):
            d[i,j] =  array[i-1,j] + array[i+1,j] + array[i, j-1]+array[i,j+1] - 4. * array[i,j]
            
    return d * .25
```

Μια πιο αποδοτική υλοποίηση με χρήση της τεχνικής του vectorization:

```python

def laplacian_v(array):
    n = array.shape[0]
    m = array.shape[1]
    d = array[:n-2, 1:m-1] + array[2:, 1:m-1] + array[1:n-1, :m-2] + array[1:n-1, 2:] - 4. * array[1:n-1, 1:m-1]
    d *= .25
    return d
```

Η φωτεινότητα των στοιχείων της εικόνας προκύπτει αθροίζοντας τις τιμές των τριών καναλιών των χρωμάτων με κατάλληλη κλήση της συνάρτησης `sum()`. Εδώ η `sum()` αθροίζει τα στοιχεία του πίνακα img κατά την τρίτη διάσταση η οποία περιγράφει το κανάλι (axis=2).

```python

flat_image = np.sum(img, axis=2)
plt.imshow(flat_image)
plt.axis('Off')
plt.show()
```

```python
ii = laplacian_v(flat_image)
plt.imshow(ii)
plt.axis('Off')
plt.show()
```

Μπορούμε να πάρουμε μια πιο έντονη απεικόνιση, επιλέγοντας από τον πίνακα τις τιμές με φωτεινότητα πάνω από ένα όριο (την τιμή 30):

```python

plt.imshow(ii > 30)
plt.axis('Off')
plt.show()
```
