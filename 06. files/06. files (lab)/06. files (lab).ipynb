## Wordle

Το Wordle, https://www.nytimes.com/games/wordle/index.html,  είναι ένα παιχνίδι λέξεων όπου ο παίκτης προσπαθεί να μαντέψει μια λέξη 5 γραμμάτων σε 6 προσπάθειες. Μετά από κάθε προσπάθεια, το πρόγραμμα εμφανίζει πληροφορίες για το πόσο κοντά είναι η λέξη που μάντεψε ο παίκτης στη σωστή λέξη με τον εξής τρόπο:

- 🟩 Πράσινο → Το γράμμα είναι σωστό και στη σωστή θέση.
- 🟨 Κίτρινο → Το γράμμα υπάρχει στη λέξη, αλλά είναι σε λάθος θέση.
- ⬜ Γκρι → Το γράμμα δεν υπάρχει καθόλου στη λέξη.

1. Δημιουργήστε συνάρτηση, έστω _loadWords_, η οποία δέχεται το όνομα του αρχείου με τις λέξεις και δημιουργεί και επιστρέφει λίστα που περιέχει όλες τις λέξεις του αρχείου.
Παράδειγμα αρχείου:
```
apple
grape
table
chair
mouse
plant
```
Παράδειγμα κλήσης:
```
>>>loadWords("words.txt")

['apple', 'grape', 'table', 'chair', 'mouse', 'plant',...
```

2. Υλοποιήστε συνάρτηση, έστω _checkQuess_ η οποία δέχεται δύο λέξεις. Την προς αναζήτηση λέξη, έστω _secret_ και την επιλογή του παίκτη, έστω _guess_. Επιστρέφει `str` με ενδείξεις για κάθε γράμμα της λέξης _guess_ σε σχέση με την λέξη _secret_. Για κάθε γράμμα που είναι στη σωστή θέση την ένδειξη 'v' για κάθε λέξη που δεν υπάρχει την ένδειξη 'x' και για κάθε λέξη που υπάρχει αλλά σε άλλη θέση την ένδειξη '-'. Προσοχή: Στο επόμενο παράδειγμα το γράμμα 'n' υπάρχει στην λέξη 'snail' αλλά μόνο μία φορά. Έτσι Στην 1η παρουσία του 'n' στην λέξη 'sunny' εμφανίστηκε η ένδειξη '-' γιατί υπάρχει αλλά στην λάθος θέση. Στην 2η παρουσία του 'n' θα πρέπει να εμφανιστεί η ένδειξη 'x' γιατί δεν υπάρχει 2η φορά στην 'snail'.

Παράδειγμα κλήσης:
```

```


3. Υλοποιήστε συνάρτηση, έστω  _wordleGame_, η οποία υλοποιεί την λογική του παιχνιδιού ως εξης:

        - Ο παίκτης θα πρέπει να ερωτάται για την λέξη που μαντεύει, έστω _guess_, και ελέγχεται αν η λέξη υπάρχει στη λίστα λέξεων και αν η μυστική λέξη έχει βρεθεί.

        - Μετά από κάθε προσπάθεια εμφανίζεται το αποτέλεσμα κάθε προσπάθειας καλώντας κατάλληλα την _checkQuess_.
