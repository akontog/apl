# Μεταγλώττιση - Διερμήνευση (Compilation - Interpretation)

Μεταγλώττιση, **Compilation** -- ο πηγαίος κώδικας μεταφράζεται σε γλώσσα μηχανής μία φορά (ωστόσο, αυτή η πράξη πρέπει να επαναλαμβάνεται κάθε φορά που τροποποιείτε τον πηγαίο κώδικα) και παράγεται ένα ανεξάρτητο εκτελέσιμο αρχείο π.χ. *.exe* για Windows μηχανήματα.

-   Το παραγόμενο αρχείο μπορεί να "τρέξει" σε μηχανήματα με παρόμοια αρχιτεκτονική και λειτουργικό σύστημα.

-   Tο πρόγραμμα που εκτελεί αυτή τη μετάφραση ονομάζεται μεταγλωττιστής ή μεταφραστής, **compiler**.

Διερμήνευση, **Interpretation** -- Στις διερμηνευόμενες γλώσσες προγραμματισμού το πηγαίο πρόγραμμα μεταφράζεται σε γλώσσα μηχανής κάθε φορά που πρέπει να εκτελεστεί.

-   Ο τελικός χρήστης χρειάζεται επίσης τον διερμηνέα για να \"τρέξει\" εκτελέσει τον πηγαίο κώδικα.

-   Το πρόγραμμα που εκτελεί αυτό το είδος μετασχηματισμού ονομάζεται διερμηνέας, **interpreter**, καθώς ερμηνεύει τον κώδικα κάθε φορά που πρόκειται να εκτελεστεί.

# Βασικοί Τύποι Δεδομένων

Κάθε τιμή στην Python έχει έναν *τύπο*, `type`, που υποδηλώνει το είδος των δεδομένων που αντιπροσωπεύει. Ο παρακάτω πίνακας περιγράφει τους βασικούς τύπους της γλώσσας με αντίστοιχα παραδείγματα.

  |Τύπος|              Παραδείγματα|
  |:--------|:---------|
  | **Απλοί Τύποι**                    | |
 | Ακέραιοι, `int`                    |                 15, -9, 1026614471576574947|
 | Πραγματικοί (αριθμοί κινητής υποδιαστολής), `float`|   3.14159, 100.0, 314159E-5|
 | Μιγαδικοί, `complex`               |                     3+5j|
 | Λογικές τιμές, `bool`         |                       `True`, `False`|
 | `NoneType`         |                       `None`|

 | **Σύνθετοι Τύποι**                    | |
 | Λίστες, `list`           |                            [1, 2, [3, 2]], [1, 1.5, “Nikos”]|

## `round`

Όταν η `int` εφρμόζεται σε τύπου `float`, σταθερά ή μεταβλητή, δεν γίνεται στρογγυλοποίηση αλλά αποκοπή του δεκαδικού μέρους.

````python
int(6.71)
````

Αν θέλουμε στρογγυλοποίηση χρησιμοποιούμε την `round`.

````python
round(6.71)
````

Θα πρέπει να δοθεί προσοχή όταν στρογγυλοποιούμε πραγματικό με δεκαδικό μέρος το 0.5. Η στρογγυλοποίηση γίνεται προς το πλησιέστερο άρτιο. Μία διαδικασία που ονομάζεται στρογγυλοποίηση τραπεζίτη.

````python
round(4.5), round(5.5)
````

Η `round` μπορεί να δεχτεί μία δεύτερη παράμετρο που καθορίζει τον αριθμό του δεκαδικού ψηφίου στο οποίο θα γίνει η στρογγυλοποίηση.

````python
round(3.14159,2)
````

## Αριθμητικοί τελεστές 

Οι παρακάτω τελεστές χρησιμοποιούνται για την
εκτέλεση αριθμητικών πράξεων:


| Τελεστής | Λειτουργία |
|:---------:|:------------: |
| `**` | Ύψωση σε δύναμη |
| `*` `/` `//` `%` | Πολλαπλασιασμός, Διαίρεση, Ακέραια διαίρεση, Υπόλοιπο (Modulo) |
| `+` `-` | Πρόσθεση, Αφαίρεση |


````python
result = 1 + 1;          # result is 2 
result = result - 1;     # result is now 1
result = result * 3;     # result is now 3 
result = result // 2;    # result is 1 (!) 
result = result + 8;     # result is now 9 
print(result)
````

### Ακρίβεια

Προσοχή πρέπει να δοθεί στις πράξεις μεταξύ πραγματικών αριθμών. Το αποτέλεσμα της έκφρασης `1.1 + 2.2` στην python δεν θα είναι ακριβώς `3.3` όπως θα περίμενε κανείς, αλλά κάτι σαν `3.30000000000000003`. Αυτό συμβαίνει λόγω της φύσης του τρόπου που οι υπολογιστές αναπαριστούν αριθμούς κινητής υποδιαστολής, **floating-point numbers**. Σε περίπτωση που η ακρίβεια είναι σημαντική θα πρέπει να γίνει χρήση της βιβλιοθήκης της python `decimal` η οποία αποθηκεύει τους αριθμούς στο δεκαδικό σύστημα.

````python
result = 1.1 + 2.2
result
````

````python
import decimal
result = decimal.Decimal('1.1')+decimal.Decimal('2.2')
result
````

## Δυαδικοί (bitwise) τελεστές

H Python επιτρέπει τις πράξεις μεταξύ _ακεραίων_ στο επίπεδο bit των αριθμών. Οι τελεστές αυτοί είναι ακριβώς ίδιοι όπως και στη γλώσσα C.

| &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;  &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; |     |
|:---------:|:------------|
| `&` | Εκτελεί τη πράξη **ΚΑΙ**, **AND**, στα αντίστοιχα bit των δύο μελών, δηλαδή επιστρέφει 1 όταν και τα 2 bit είναι 1 αλλιώς επιστρέφει 0 |
| `\|` | Εκτελεί τη πράξη **Ή**, **OR**, στα αντίστοιχα bit των δύο μελών, δηλαδή επιστρέφει 1 όταν ένα από τα 2 bit είναι 1 αλλιώς επιστρέφει 0 |
| `^` | Εκτελεί τη πράξη **αποκλειστικό Ή**, **XOR** στα αντίστοιχα bit των δύο μελών, δηλαδή επιστρέφει 1 όταν μόνο ένα από τα 2 bit είναι 1 αλλιώς επιστρέφει 0 |
| `~` | Αντιστρέφει τα bits, δηλαδή τα 1 γίνονται 0 και τα 0 γίνονται 1 |
| `<< n` | Μετακινεί τα bits n-θέσεις αριστερά και τα κενά συμπληρώνονται με 0 |
| `>> n` | Μετακινεί τα bits n-θέσεις δεξιά και τα κενά συμπληρώνονται με 0 |

Η συνάρτηση εξόδου `print` Καθώς και οι μορφοποιημένες συμβολοσειρές θα εξεταστούν αμέσως μετά.

````python
a = 42
print (f"Δεκαδική αναπαράσταση:{a:2} Δυαδική αναπαράσταση:{a:8b}")
b = 22
print (f"Δεκαδική αναπαράσταση:{b:2} Δυαδική αναπαράσταση:{b:8b}")
````

Θέτοντας αρχικά `0b` σε κάποιον αριθμό δηλώνουμε ότι πρόκεται για δυαδική αναπαράσταση του αριθμου.

````python
0b101010,0b10110
````

````python
print ("Bitwise operator: AND ")
c = a & b
print (f"{a}&{b} -> Δεκαδική αναπαράσταση:{c:2} Δυαδική αναπαράσταση:{c:8b}")
````

````python
c = 0b101010 & 0b10110
c,bin(c)
````

````python
print ("Bitwise operator: OR ")
c = a | b
print (f"{a}|{b} -> Δεκαδική αναπαράσταση:{c:2} Δυαδική αναπαράσταση:{c:8b}")
````

````python
c = 0b101010 ^ 0b10110
c,bin(c)
````

````python
print ("Bitwise operator: XOR ")
c = a ^ b
print (f"{a}^{b} -> Δεκαδική αναπαράσταση:{c:2} Δυαδική αναπαράσταση:{c:8b}")
````

````python
print ("Bitwise operator: NOT ")
c = ~ a
print (f" ~ {a} -> Δεκαδική αναπαράσταση:{c:2} Δυαδική αναπαράσταση:{c:8b}")
````

````python
print ("Bitwise operator: >> n ")
c = a >> 2
print (f"{a} >> 2 -> Δεκαδική αναπαράσταση:{c:2} Δυαδική αναπαράσταση:{c:8b}")
````

````python
print ("Bitwise operator: << n ")
c = a << 2
print (f"{a} << 2 -> Δεκαδική αναπαράσταση:{c:2} Δυαδική αναπαράσταση:{c:8b}")
````

## Τελεστές ταυτότητας, identity, `is`

Ο τελεστής ταυτότητα `is` ελέγχει αν οι τελεστέοι αναφέρονται στο ίδιο αντικείμενο. 
Προσοχή:

 - Ακέραιοι στο [-5,256] αποθηκεύονται και επαναχρησιμοποιούνται.

- Τα αλφαριθμητικά αποθηκεύονται στη μνήμη και για όλα όσα έχουν με την ίδια τιμή, η Python θα χρησιμοποιήσει το ίδιο αντικείμενο από το "literal pool" αντί να δημιουργήσει ένα νέο.

Διαφορά μεταξύ `==`, `is`:

- Ο τελεστής `==` ελέγχει αν τα αντικείμενα έχουν την ίδια τιμή.

- Ο `is` αν οι τελεστέοι αναφέρονται στο ίδιο αντικείμενο.

````python
a = 276
b = 276
print (f"{a:3} is {b:3}:\t{a is b}")
print (f"{a:3} is not {b:3}:\t{a is not b}")
a = 5
b = 5
print (f"{a:3} is {b:3}:\t{a is b}")    #  [-5,256]
print (f"{a:3} is not {b:3}:\t{a is not b}")
# Λίστες, θα αναφερθούν στην συνέχεια
L1 = [1,2,3]
L2 = [1,2,3]
L3 = L1 
print(f"L1==L2:\t\t{L1==L2}")     # ίδιο περιεχόμενο
print(f"L1 is L2:\t{L1 is L2}")   # διαφορετικό αντικείμενο
print(f"L1==L3:\t\t{L1==L3}")     # ίδιο περιεχόμενο
print(f"L1 is L3:\t{L1 is L3}")   # ίδιο αντικείμενο
s1 = "Samos"
s2 = "Samos"
s3 = s1
print (f"s1 is s2:\t{s1 is s2}")
print (f"s1==s2:\t\t{s1==s2}")     # literal pool
print (f"s2 is s3:\t{s1 is s3}")
print (f"s2==s3:\t\t{s1==s3}")
````

## Προτεραιότητα τελεστών

Οι τελεστές στη Python έχουν προτεραιότητα όμοια με αυτή πολλών άλλων γλωσσών προγραμματισμού.

-   Υψηλότερη προτεραιότητα έχει η δύναμη `**` και έπονται οι μονομελείς τελεστές. Ακολουθούν οι αριθμητικοί τελεστές, οι τελεστές bit (με σειρά προτεραιότητας: {`>>` ή `<<`}, `&`, `^`, `|`), οι τελεστές συσχέτισης και τέλος οι τελεστές ανάθεσης.

-   Η προτεραιότητα των αριθμητικών τελεστών και των τελεστών συσχέτισης είναι από αριστερά προς τα δεξιά.

-   Η προτεραιότητα των τελεστών ανάθεσης από δεξιά προς τα αριστερά.

-   Οι τελεστές πολ/μού και διαίρεσης έχουν υψηλότερη προτεραιότητα από αυτούς της πρόσθεσης και της αφαίρεσης.

Για την αποφυγή σφαλμάτων, καλό είναι να χρησιμοποιούνται παρενθέσεις.


| **Κατηγορία**          | **Τελεστής**                                                             | **Προσεταιριστικότητα** |
|-------------------------|-------------------------------------------------------------------------|-------------------------|
| **Αριθμητικοί**         | **\*\***                                                              | **←**                  |
|                         | (μονομελείς) `+x`, `-x`, `~x`                                                | **←**                  |
|                         | `*`, `/`, `//`, `%`                                                           | **→**                  |
|                         | `+`, `-`                                                                 | **→**                  |
| **Δυαδικοί**            | `<<`, `>>`                                                                | **→**                  |
|                         | `&`                                                                     | **→**                  |
|                         | `^`                                                                     | **→**                  |
|                         | |                                                                     | **→**                  |
| **Σχεσιακοί, Ταυτότητας** | `==`, `!=`, `>`, `<`, `>=`, `<=`, `is`, `is not`                                       | **→**                  |
| **Λογικοί**             | `not`                                                                   | **←**                  |
|                         | `and`                                                                   | **→**                  |
|                         | `or`                                                                    | **→**                  |

## Δεικτοδότηση (indexing), τεμαχισμός (slicing)

Όλοι οι ακολουθιακοί τύποι της Python επιτρέπουν την προσπέλαση των στοιχείων τους με τη χρήση αριθμόδεικτών, _index_, ο οποίος προσδιορίζει τη θέση του στοιχείου μέσα στην ακολουθία. Η αρίθμηση μπορεί να ξεκινά είτε από τα αριστερά και το 0 είτε από τα δεξιά με τον αριθμοδείκτη -1 να παριστάνει το τελευταίο στοιχείο.
    
 
|  0  |  1  |  2  |  3  |  4  |  5  |
|:---:|:---:|:---:|:---:|:---:|:---:|
| p   | y   | t   | h   | o   | n   |
| -6  | -5  | -4  | -3  | -2  | -1  |


````python
text = 'Hello World'
type(text)
````

````python
print(text[0], text[6], text[-1]) # 
````

Το **slicing** στην Python είναι ένας τρόπος για να εξαγάγουμε ένα υποσύνολο από χαρακτήρες ενός string. Υλοποιείται χρησιμοποιώντας τη σύνταξη του τελεστή τεμαχισμού (slice operator) με την μορφή: string[start:stop:step].

- **start**: Ο δείκτης, index, από τον οποίο ξεκινάει το slicing και περιλαμβάνεται. Αν δεν υπάρχει, υπονοείται το 0 η αρχή του string.
- **stop**: Ο δείκτης στον οποίο σταματάει το slicing χωρίς να περιλαμβάνεται. Αν δεν υπάρχει, υπονοείται το len(string) το τέλος του string.
- **step**:Το βήμα με το οποίο προχωράει το slicing. Αν δεν υπάρχει, υπονοείται το 1.

````python
text = 'Hello World'
print(text[2:8])
print(text[:5])
print(text[:])
````

````python
text = 'Hello World'
print(text[:-3])
print(text[3:-3])
````

````python
text = 'Hello World'
print(text[::2])  # Τεμαχισμός με βήμα
````

## Διαίρεση και συνένωση, split, join

Η μέθοδος `split()`, ξεχωρίζει τις λέξεις μέσα σε μια συμβολοσειρά με τη βοήθεια της διαχωριστικής συμβολοσειράς του ορίσματος της. Αν δεν δοθεί κάποιο όρισμα η `split()` χρησιμοποιεί το κενό. Μπορεί να δοθεί και ένα δεύτερο όρισμα που καθορίζει πόσες φορές θα γίνει διαχωρισμός.

````python
text = 'Hello World'
newText = text.split()
print(newText)
````

````python
text = 'Hello World'
newText = text.split('o')
print(newText)
````

````python
text = 'Hello World'
newText = text.split('o',1)
print(newText)
````

Η μέθοδος `join()` συνενώνει τις συμβολοσειρές του ορίσματος της σε μια. Το διαχωριστικό μεταξύ των στοιχείων είναι η συμβολοσειρά επι της οποίας καλείται η `join()`.

````python
origText = ' '.join(newText)
print(origText)
````

Μέσα στα άγκιστρα \{ \}, μπορούν να οριστούν τα δεκαδικά εμφάνισης του αριθμού, το πλάτος του πεδίου εμφάνισης του αριθμού, η στοίχιση του κ.ο.κ.

Κάποια από τα πιο χρησιμοποιούμενα σύμβολα μορφοποίησης είναι:

|Σύμβολο μορφ. |Λειτουργία |
|:---------:|:------------ |
| `:<`  |  Αριστερή στοίχιση μέσα στον διαθέσιμο χώρο |
| `:>`  |  Δεξιά στοίχιση μέσα στον διαθέσιμο χώρο |
| `:^`  |  Στοίχιση στο κέντρο μέσα στον διαθέσιμο χώρο |
| `:+`  |  Εμφανίζει το πρόσημο μπροστά από τους αριθμούς |
| `:,`  |  Χωρίζει τις χιλιάδες με κόμμα |
| `:c`  |  Εμφανίζει τον αντίστοιχο Unicode χαρακτήρα |
| `:b`  |  Εμφανίζει τη δυαδική μορφή ενός ακέραιου |
| `:x`  |  Εμφανίζει τη δεκαεξαδική μορφή ενός ακέραιου |
| `:e`  |  Εμφανίζει έναν αριθμό σε επιστημονική (εκθετική) μορφή |

**Παράδειγμα**

Η τιμή του π στις επόμενες εκτυπώσεις θα στοιχιστεί, εξορισμού, δεξιά, αριστερά και στο κέντρο. Χρησιμοποιείται η `print` που θα αναλυθεί αμέσως μετά καθώς και η μαθηματική βιβλιοθήκη `math` που περιέχει διάφορες χρήσιμες ιδιότητες και μεθόδους.

````python
import math # Εισάγεται η μαθηματική βιβλιοθήκη, περισσότερα σε επόμενη ενότητα
print(f'The value of pi is approximately {math.pi:20.10f}!')
print(f'The value of pi is approximately {math.pi:<20.10f}!')
print(f'The value of pi is approximately {math.pi:^20.10f}!')
````

````python
print('Unicode char:\t{0:c} \nBinary form:\t{0:b} \
        \nShow sign:\t{0:+}\t{1:+} \
      \nE notation:\t{1:e}'.format(65, -12.563))
````
